# 🧹 Notebook 2 — Text Preprocessing
Remove noise, stopwords, lemmatize, and save cleaned data.

```python
import sys, os
sys.path.append('..')
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from src.preprocess import clean_text, preprocess_dataframe, save_processed

print('✅ Imports done!')
```

```python
# Load raw data
train_df = pd.read_csv('../data/raw/train.csv')
test_df  = pd.read_csv('../data/raw/test.csv')
print('Loaded:', train_df.shape, test_df.shape)
```

```python
# Demo clean_text
sample = train_df['text'].iloc[0]
print('ORIGINAL:')
print(sample)
print('\nCLEANED:')
print(clean_text(sample))
```

```python
# Preprocess full datasets
train_df = preprocess_dataframe(train_df)
test_df  = preprocess_dataframe(test_df)
train_df[['text','clean_text','text_length','clean_length']].head()
```

```python
# Compare lengths before/after
fig, axes = plt.subplots(1,2, figsize=(12,4))
sns.histplot(train_df['text_length'],  bins=50, ax=axes[0], color='steelblue').set_title('Before Cleaning')
sns.histplot(train_df['clean_length'], bins=50, ax=axes[1], color='#55A868').set_title('After Cleaning')
for ax in axes:
    ax.set_xlabel('Word Count')
plt.tight_layout()
plt.show()
```

```python
# Save processed data
save_processed(train_df, test_df)
print('✅ Step 2 complete!')
```
