# 📊 Notebook 1 — EDA & Dataset Loading
**Step 1:** Load AG News, explore structure, visualize class distribution and text lengths.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datasets import load_dataset
import os

print('✅ Libraries imported!')
```

```python
# Load dataset
print('⏳ Loading AG News...')
dataset = load_dataset('ag_news')
print(dataset)
```

```python
# Convert to pandas
train_df = pd.DataFrame(dataset['train'])
test_df  = pd.DataFrame(dataset['test'])

label_names = {0:'World', 1:'Sports', 2:'Business', 3:'Technology'}
train_df['category'] = train_df['label'].map(label_names)
test_df['category']  = test_df['label'].map(label_names)

print('Train:', train_df.shape)
print('Test :', test_df.shape)
train_df.head()
```

```python
# Class distribution
print('Train class counts:')
print(train_df['category'].value_counts())
print('\nTest class counts:')
print(test_df['category'].value_counts())
```

```python
# Plot class distribution
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
colors = ['#4C72B0','#DD8452','#55A868','#C44E52']

train_df['category'].value_counts().plot(kind='bar', ax=axes[0], color=colors, edgecolor='black')
axes[0].set_title('Train Set - Class Distribution', fontsize=13)
axes[0].tick_params(axis='x', rotation=45)

test_df['category'].value_counts().plot(kind='bar', ax=axes[1], color=colors, edgecolor='black')
axes[1].set_title('Test Set - Class Distribution', fontsize=13)
axes[1].tick_params(axis='x', rotation=45)

os.makedirs('../outputs/plots', exist_ok=True)
plt.tight_layout()
plt.savefig('../outputs/plots/class_distribution.png', dpi=150)
plt.show()
print('✅ Plot saved!')
```

```python
# Text length analysis
train_df['text_length'] = train_df['text'].apply(lambda x: len(x.split()))
print(train_df['text_length'].describe().round(2))

plt.figure(figsize=(10,4))
sns.histplot(train_df['text_length'], bins=50, color='steelblue', kde=True)
plt.axvline(train_df['text_length'].mean(), color='red', linestyle='--', label=f"Mean: {train_df['text_length'].mean():.1f}")
plt.title('Word Count per Sample')
plt.xlabel('Number of Words')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Sample texts per category
for cat in label_names.values():
    sample = train_df[train_df['category']==cat]['text'].iloc[0]
    print(f'\n📌 {cat}:')
    print(sample[:200], '...')
    print('-'*60)
```

```python
# Save raw data
os.makedirs('../data/raw', exist_ok=True)
train_df.to_csv('../data/raw/train.csv', index=False)
test_df.to_csv('../data/raw/test.csv', index=False)
print('✅ Saved to data/raw/')
```
